```python
from facenet_pytorch import MTCNN
import cv2
from PIL import Image
from google.colab.patches import cv2_imshow

# Suponhamos que esta lista tenha os nomes dos seus funcionários
nomes_funcionarios = ["WILMAR", "LARISSA", "ANDERSON", "ORDILEY", "FABIO SILVA", "EVANDRO MARCOS", "TIAGO FRANCA", "IGOR SEG TRABALHO ", "MARCIO COIADA","RICARDO","FABRICIO"]

# Carregar a imagem
imagem = cv2.imread('/content/7.jpeg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)  # Converta a imagem para RGB
imagem_pil = Image.fromarray(imagem_rgb)  # Converta a imagem para o formato PIL

# Inicializar o detector MTCNN
mtcnn = MTCNN(keep_all=True)

# Detectar rostos
boxes, probs = mtcnn.detect(imagem_pil)

# Desenhar retângulos ao redor das faces e adicionar nomes
for i, box in enumerate(boxes):
    cv2.rectangle(imagem, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (0, 255, 0), 2)
    nome = nomes_funcionarios[i] if i < len(nomes_funcionarios) else f'Desconhecido {i + 1}'
    cv2.putText(imagem, nome, (int(box[0]), int(box[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Exibir a imagem
cv2_imshow(imagem)
```

```python
import cv2
from facenet_pytorch import MTCNN
from PIL import Image
from google.colab.patches import cv2_imshow

# Suponhamos que esta lista tenha os nomes dos seus funcionários
nomes_funcionarios = ["WILMAR", "LARISSA", "ANDERSON", "ORDILEY", "FABIO SILVA", "EVANDRO MARCOS", "TIAGO FRANCA", "IGOR SEG TRABALHO ", "MARCIO COIADA","RICARDO","FABRICIO"]

# Carregar a imagem
imagem = cv2.imread('/content/7.jpeg')

# Redimensionar a imagem
h, w = imagem.shape[:2]  # Obter as dimensões da imagem original
largura_desejada = 900  # Definir a largura desejada
altura_desejada = int((largura_desejada / w) * h)  # Calcular a altura proporcional
imagem = cv2.resize(imagem, (largura_desejada, altura_desejada))  # Redimensionar a imagem

# Converter a imagem para RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
imagem_pil = Image.fromarray(imagem_rgb)

# Inicializar o detector MTCNN
mtcnn = MTCNN(keep_all=True)

# Detectar rostos
boxes, probs = mtcnn.detect(imagem_pil)

# Desenhar retângulos ao redor das faces e adicionar nomes
for i, box in enumerate(boxes):
    cv2.rectangle(imagem, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (0, 255, 0), 2)
    nome = nomes_funcionarios[i] if i < len(nomes_funcionarios) else f'Desconhecido {i + 1}'
    cv2.putText(imagem, nome, (int(box[0]), int(box[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Exibir a imagem
cv2_imshow(imagem)
```
